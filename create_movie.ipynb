```python
import os
from bs4 import BeautifulSoup



root_folder = '/home/ankit/Desktop/AI_course/'
alldirs = os.listdir(root_folder)
all_dirs = [direc for direc in alldirs if direc.startswith('Part')]

```

```python
import pandas as pd 

main_df =[]
topic_list=[]
part_list=[]
video_list=[]


    
for dir in sorted(all_dirs):
    with open('./lessions.md','w') as fout:
            
        id=dir.replace(' ','_')
        id = id.replace('.','_')
        
        fout.write(f'## {dir}\n\n')
        print(dir)
        # get file list
        filelist = os.listdir(f'{root_folder}{dir}')
        # get videos
        videos = [f for f in filelist if f.endswith('.mp4')]
        vidno = [vid.split('.')[0] for vid in videos]
        # print('-----')
        with open(f'{root_folder}{dir}/index.html', 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')

        
            for li_tag in soup.find_all('li'):
                # Find all <a> tags within the <li> tag
                for a_tag in li_tag.find_all('a', href=True):
                    topic = a_tag['href']
                    topich = topic.replace('.html','')
                    fout.write(f'### {topich}\n')
                    
                    if topic.split('.')[0] in vidno:
                        
                        filename = f"{root_folder}{dir}/{videos[vidno.index(topic.split('.')[0])]}"

                        
                        fout.write(f"""<video width="600" controls> <source src="{filename}" type="video/mp4"></video>\n\n""")
                     
                        
                    else:

                        filename = f'{root_folder}{dir}/{topic}'.replace(' ','_')
                        
                        fout.write(f"""<a href="{filename}" target="_blank">{topic}</a> \n\n""")

                
        # fout.write('</section>\n')
    
```
