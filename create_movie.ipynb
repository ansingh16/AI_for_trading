```python
import os
from bs4 import BeautifulSoup


def find_all_index_files(root_folder):

    all_index=[]    
    all_videos=[]
    # Walk through the folder and subfolders
    for subdir, dirs, files in os.walk(root_folder):
        
        for file in files:
            if file == 'index.html':
                index_path = os.path.join(subdir, file)

                all_index.append(index_path)

            if file.endswith('.mp4'):
                all_videos.append(file)

        
    return all_index,all_videos


if __name__ == "__main__":
    root_folder = "."  # Replace with your root folder path
    all_index,all_videos = find_all_index_files(root_folder)
    all_dirs = [link.split('/')[1] for link in all_index]
    # # Output the links
    # for folder, links in all_video_links.items():
    #     print(f"Video links in folder: {folder}")
    #     for link in links:
    #         print(f"  - {link}")
```
